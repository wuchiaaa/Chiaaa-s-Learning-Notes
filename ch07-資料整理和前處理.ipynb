### 離散化和分組

```python
'''
cut
'''
ages = [20,22,25,27,21,23,37,31,61,45,41,32, 101]
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)  # 默認 right=True: 包含右邊
cats
```

```python
cats.codes  # 原labels已棄用
```

```python
cats.categories  # 原levels已棄用
```

```python
pd.value_counts(cats)  # 分組以後每組數量
```

```python
group_name = ['Youth', 'YoughAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_name)
```

```python
data = np.random.rand(20)  # rand:的隨機樣本位於[0, 1)中 ; randn:從標準正態分布中返回一或多個樣本值
data
```

```python
pd.cut(data, 4, precision=2)  # bins=4, precision=2 限定小數點下兩位
```

```python
'''
qcut
'''
data = np.random.randn(1000) # 常態分布
cats = pd.qcut(data, 4) # 切成四分，同[0, .25, .5, .75, 1]
cats
```

```python
pd.value_counts(cats)
```

```python
cats2 = pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])  # 自訂每組的數量，範圍0~1(包含1)
cats2
```

```python
pd.value_counts(cats2)
```

```python
'''
來源: GroupLens Reasearch (https://grouplens.org/datasets/movielens/)
取用 MovieLens 1M 資料集合
''' 
import pandas as pd
import numpy as np
mnames = ['movie_id', 'title', 'genres']
```

```python
movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames)
```

```python
movies[:10]
```

```python
# 針對電影分類(genres)做指標變數
all_genres = []

for x in movies.genres:
    all_genres.extend(x.split('|'))
    
genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
```

```python
zero_matrix.size
```

```python
dummies = pd.DataFrame(zero_matrix, columns=genres)
```

```python
dummies
```

```python
# 在對應分類上標記 1 (dummies的列上項目)
gen = movies.genres[1]
gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))  # 分類的index
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

```python
# 把結果和movies合併
movies_windic = movies.join(dummies.add_prefix('Genre_'))
```

```python
movies_windic.iloc[0]
```

```python
movies_windic
```

### 正規表達式

```python
import re
text = "foo   bar\t baz  \tqux"
re.split('\s+', text)
```

```python
regex = re.compile('\s+')
regex
```

```python
regex.split(text)
```

```python
text = """Dave dave@gmail.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@gmail.com
"""

pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
```

```python
regex = re.compile(pattern, flags=re.IGNORECASE)   #IGNORECASE:可使 regex 不管大小寫
```

```python
regex.findall(text)
```

```python
# sub:取代
print(regex.sub('REDACTED', text))
```

```python
# 找到郵件帳號，並切割成 使用者名稱、網域名稱、網域後綴 > 放括號
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern, flags=re.IGNORECASE)
```

```python
m = regex.match('wesm@brightt.net')   # 回傳match物件
m.groups()  # 得到分段好的元素
```

```python
regex.findall(text)
```
